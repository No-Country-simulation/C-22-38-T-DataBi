```python
#Cargamos Librerias
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler 
from sklearn.decomposition import PCA 
from scipy import stats

```

```python
#Cargamos el data
data = pd.read_csv ('dataset_phishing.csv')
```

```python
# verificamos si hay filas duplicadas
duplicados= data.duplicated()
print(duplicados.sum())
```

```python
# Deteccion y tratamiento de outliers usando Z-score
z_scores=np.abs(stats.zscore(data.select_dtypes(include=[np.number])))# calcula los z-scores para todas las columnas-np.abs para obtener los valores absolutos
outliers = data[(z_scores < 3).all(axis=1)] #selecciona las filas donde cualquier columna numerica tiene un z-score
print(outliers)#imprime la filas que son outliers
#Un valor considerado un outliers si su Z-score es mayor q 3 o menor q -3.
#El Z-score mide cuantas desviaciones estandar esta un dato de la media.
```

```python
#normalizacion usando Min-Max Scaling
#creamos variable 
scaler= MinMaxScaler()
#seleccion de columnas numericas
numerical_columns= data.select_dtypes(include=[np.number]).columns
# Opcional: Rellena los valores NaN con la mediana de cada columna
data[numerical_columns] = data[numerical_columns].fillna(data[numerical_columns].median())

#aplicamos el escalado Min-Max
data[numerical_columns]=scaler.fit_transform(data[numerical_columns])

#guardamos el data normalizado en un nuevo archivo CSV
data.to_csv ('data_normalizado.csv', index=False)

print(data.head())

#La normalizacion es una tecnica para ajustar los valores en un rango especifico, generalmente entre 0 y 1. 

```

```python
#Metodo:seleccion de caracteristicas basada en la varianza
# Eliminacion de columnas que presentan baja variabilidad
# Calculamos la varianza de las columnas

#seleccionamos las columnas numericas
numerical_columns=data.select_dtypes(include=[np.number])

#calculamos la varianza
varianzas=numerical_columns.var()

#identificamos columnas con baja varianza
threshold = 0.00020 #definimos el valor de threshold para eliminar aquellas columnas que tenga la varianza menos q este valor
columnas_varianza_baja= [column for column in varianzas.index if varianzas [column]< threshold]
 
 #mostramos columnas con baja varianza
print("las siguientes columnas tiene baja varianza:")
print((columnas_varianza_baja))
```
